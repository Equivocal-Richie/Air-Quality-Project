<h1>This notebook helps to understand the data before working on it eg. data cleaning</h1>

```python
# import necessary libraries
import pandas as pd
import sqlite3
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data from the sqlite database
def load_data_from_sqlite(db_path):
    conn = sqlite3.connect(db_path)
    query = "SELECT * FROM air_quality"
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df

# Get the current working directory
current_dir = os.getcwd()

# Construct the database path relative to the current working directory
db_path = os.path.join(current_dir, "..", "data", "raw", "air_quality_data.db")

df = load_data_from_sqlite(db_path)
```

```python
# Get the statistical summary of the numerical columns
df.describe()
```

<h2> <b> Key Observations from Initial Inspection: </b></h2>

1. Missing Values:
   - Columns o3, no2, so2, and co have 0 non-null counts, meaning they are completely empty. This is a critical issue.
2. Data Types:
   - timestamp is an object (string) and needs to be converted to a datetime object.
   - All numerical columns (latitude, longitude, aqi, pm25, pm10, temperature, humidity, wind_speed, wind_direction, pressure) are float64, which is generally appropriate.
3. Unique ID Column:
   - An id column exists. Since it's just an index, it's redundant and can be removed.
4. Descriptive Statistics:
   The statistical summary (df.describe()) provides insights into the distributions of numerical columns.
   - aqi, pm25, and pm10 have varying ranges.
   - wind_direction ranges from 0 to 360, which is expected.
   - pressure has a reasonable range, but it's worth checking for outliers.
5. Memory Usage:
   - The memory usage (956.1+ KB) is relatively low, indicating that the dataset is not excessively large.
6. Timestamp:
   - The Timestamps all appear to be in UTC time.

```python
# Visualize outliers in the numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

<h3><b>Analysis of the Box Plots:</b></h3>
<p style="font-size:18px">
    
1. aqi (Air Quality Index):
    
    - Significant number of outliers on the higher end.
    - The distribution is right-skewed, indicating that there are more instances of higher AQI values.
    - This is common in air quality data, where pollution spikes can occur.
    
2. pm25 (Particulate Matter 2.5):
    
    - Also shows a significant number of high-end outliers.
    - Similar to aqi, it's right-skewed, indicating occasional high PM2.5 concentrations.
    
3. pm10 (Particulate Matter 10):
    
    - Similar pattern to aqi and pm25, with many high-end outliers and right-skewness.
      
4. temperature:
    
    - Appears to have a relatively normal distribution with a few outliers at both the lower and higher ends.
    - The outliers might represent extreme weather conditions.
      
5. humidity:

    - Shows a fairly uniform distribution with a few outliers on the lower end.
    - Humidity values are typically bounded between 0 and 100, so the outliers are likely genuine low humidity readings.

6. wind_speed:

    - Has a right-skewed distribution with several high-end outliers.
    - These outliers might represent strong wind events or measurement errors.

7. wind_direction:

    - Appears to be uniformly distributed across the 0-360 degree range, which is expected.
    - No significant outliers are visible.

8. pressure:

    - Shows a relatively normal distribution with a few low-end outliers.
    - These outliers might represent low-pressure weather systems or measurement errors.  
</p>
